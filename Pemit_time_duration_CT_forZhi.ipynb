```python
import warnings
warnings.filterwarnings("ignore")

import requests
import numpy as np
import pandas as pd
import geopandas as gpd # GeoPandas library for spatial analytics
from shapely.geometry import shape
from shapely.geometry import shape

from datetime import datetime, timedelta
```

```python
data = gpd.read_file("../output/film permit_city.shp" )
```

```python
data_fil=data.drop(columns=['OBJECTID','Join_ID'])
data_fil=data_fil.drop_duplicates().reset_index()
```

```python
len(data_fil)/len(data)
```

```python
def range_to_list(a, b):
    return pd.date_range(a, b, freq = 'H')[:-1]
data_fil['StartEnd'] = [[i, j] for i, j in zip( data_fil['StartDateT'], data_fil['EndDateTim'])]
data_fil['TimeList'] = data_fil['StartEnd'].apply(lambda x: range_to_list(x[0], x[1]))
data_fil['hours'] = data_fil['TimeList'].apply(lambda x: list(x.hour))
data_fil['month'] = data_fil['TimeList'].apply(lambda x: list(set(x.month)))
data_fil['date'] = data_fil['TimeList'].apply(lambda x: list(set(x.date)))
data_fil['hour_sum'] = data_fil['hours'].apply(lambda x: len(x))  
data_fil['day_sum']=data_fil['hour_sum']/24
```

```python
data_fil
```

```python
data_fil['StartDateTime'] = data_fil['StartDateT'].apply(pd.to_datetime)
```

## 2021Time

```python
data_2021 = data_fil[(data_fil['StartDateTime']>='2021-1-1') & (data_fil['StartDateTime']<'2022-1-1')]
```

```python
data_2021_buffer=data_2021
data_2021_buffer['geometry']= data_2021_buffer.buffer(3,cap_style=2)
```

```python
ct = gpd.read_file("../vector/2020 Census Tract_NYC" )
ct_sel = ct[['BoroCT2020','NTAName','GEOID','geometry']]
```

```python
ct_data_2021 = ct_sel.sjoin(data_2021_buffer, how="left")
ct_data_2021=ct_data_2021.dropna().reset_index()
```

```python
month1 = ct_data_2021.apply(lambda x: pd.Series(x['month']),axis=1).stack().reset_index(level=1, drop=True)
month1.name = 'month'
ct_data_2021 = ct_data_2021.drop('month', axis=1).join(month1)
ct_data_2021['month'] = ct_data_2021['month'].astype(str)
ct_data_2021
```

```python
cd = gpd.read_file("../vector/Council Districts/nycc.shp")
cd33 = cd[cd['CounDist']==33]
ct2021_cd33 = gpd.overlay(ct_data_2021, cd33, how='intersection')
ct2021_cd33
```

```python
ct_2021_sel=ct2021_cd33[['BoroCT2020','EventID','hour_sum','day_sum','month']]
ct_2021_sel=ct_2021_sel.drop_duplicates()
ct_2021_sel['month']= ct_2021_sel['month'].astype(float).astype(int)
ct_2021_sum = ct_2021_sel[['BoroCT2020','hour_sum','day_sum','month']].groupby(by=['BoroCT2020','month']).sum()
ct_2021_sum = ct_2021_sum.reset_index()
ct_2021_sum.sort_values('day_sum')
```

```python
ct_2021_sum[ct_2021_sum['month']=='nan']
```

```python
pivot_time_2021 = ct_2021_sum.pivot_table(index =['BoroCT2020'], values =['day_sum'],columns='month')
pivot_time_2021
```
