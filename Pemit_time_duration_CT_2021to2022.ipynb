```python
import warnings
warnings.filterwarnings("ignore")

import requests
import numpy as np
import pandas as pd
import geopandas as gpd # GeoPandas library for spatial analytics
from shapely.geometry import shape
from shapely.geometry import shape

from datetime import datetime, timedelta
```

```python
data = gpd.read_file("../output/film permit_city.shp" )
```

```python
data_fil=data.drop(columns=['OBJECTID','Join_ID'])
data_fil=data_fil.drop_duplicates().reset_index()
```

```python
len(data_fil)/len(data)
```

```python
def range_to_list(a, b):
    return pd.date_range(a, b, freq = 'H')[:-1]
data_fil['StartEnd'] = [[i, j] for i, j in zip( data_fil['StartDateT'], data_fil['EndDateTim'])]
data_fil['TimeList'] = data_fil['StartEnd'].apply(lambda x: range_to_list(x[0], x[1]))
data_fil['hours'] = data_fil['TimeList'].apply(lambda x: list(x.hour))
data_fil['month'] = data_fil['TimeList'].apply(lambda x: list(set(x.month)))
data_fil['date'] = data_fil['TimeList'].apply(lambda x: list(set(x.date)))
data_fil['hour_sum'] = data_fil['hours'].apply(lambda x: len(x))  
data_fil['day_sum']=data_fil['hour_sum']/24
```

```python
data_fil
```

```python
data_fil['StartDateTime'] = data_fil['StartDateT'].apply(pd.to_datetime)
```

```python
data_2021 = data_fil[(data_fil['StartDateTime']>='2021-1-1') & (data_fil['StartDateTime']<'2022-1-1')]
```

```python
data_2021_buffer=data_2021
data_2021_buffer['geometry']= data_2021_buffer.buffer(3,cap_style=2)
```

```python
ct = gpd.read_file("../vector/2020 Census Tract_NYC" )
ct_sel = ct[['BoroCT2020','NTAName','GEOID','geometry']]
```

```python
ct_data_2021 = ct_sel.sjoin(data_2021_buffer, how="left")
ct_data_2021=ct_data_2021.dropna().reset_index()
```

```python
month1 = ct_data_2021.apply(lambda x: pd.Series(x['month']),axis=1).stack().reset_index(level=1, drop=True)
month1.name = 'month'
ct_data_2021 = ct_data_2021.drop('month', axis=1).join(month1)
ct_data_2021['month'] = ct_data_2021['month'].astype(str)
ct_data_2021
```

```python
cd = gpd.read_file("../vector/Council Districts/nycc.shp")
cd33 = cd[cd['CounDist']==33]
ct2021_cd33 = gpd.overlay(ct_data_2021, cd33, how='intersection')
ct2021_cd33
```

```python
ct_2021_forsum=ct2021_cd33[['BoroCT2020','EventID','hour_sum','day_sum','month']]
ct_2021_forsum=ct_2021_forsum.drop_duplicates()
ct_2021_forsum['month']= ct_2021_forsum['month'].astype(float).astype(int)

ct_2021_forsum
```

```python
ct_2021_sum = ct_2021_forsum[['BoroCT2020','hour_sum','day_sum','month']].groupby(by=['BoroCT2020','month']).sum()
ct_2021_sum = ct_2021_sum.reset_index()
ct_2021_sum.sort_values('day_sum')
```

```python
ct_2021_sum[ct_2021_sum['month']=='nan']
```

```python
pivot_time_2021 = ct_2021_sum.pivot_table(index =['BoroCT2020'], values =['day_sum'],columns='month')
pivot_time_2021
```

```python
pivot_time_2021.to_csv('../output/Pivot_permits_cd33_2021_time.csv')
```

```python
#2021
ct_2021_for_sum=ct_data_2021[['BoroCT2020','EventID','hour_sum','day_sum','month']]
ct_2021_for_sum=ct_2021_for_sum.drop_duplicates()
ct_2021_for_sum['month']= ct_2021_for_sum['month'].astype(float).astype(int)

ct_2021_sum_all = ct_2021_for_sum[['BoroCT2020','hour_sum','day_sum','month']].groupby(by=['BoroCT2020','month']).sum()
ct_2021_sum_all = ct_2021_sum_all.reset_index()
ct_2021_sum_all.sort_values('day_sum')

#2022
ct_2022_for_sum=ct_data_2022[['BoroCT2020','EventID','hour_sum','day_sum','month']]
ct_2022_for_sum=ct_2022_for_sum.drop_duplicates()
ct_2022_for_sum['month']= ct_2022_for_sum['month'].astype(float).astype(int)

ct_2022_sum_all = ct_2022_for_sum[['BoroCT2020','hour_sum','day_sum','month']].groupby(by=['BoroCT2020','month']).sum()
ct_2022_sum_all = ct_2022_sum_all.reset_index()
ct_2022_sum_all.sort_values('day_sum')
```

```python
ct_2021_sum_all['month_year'] = ct_2021_sum_all['month'].apply(lambda x:  '2021_'+str(x).zfill(2)) 
ct_2022_sum_all['month_year'] = ct_2022_sum_all['month'].apply(lambda x: '2022_'+str(x).zfill(2)) 
df_all = pd.concat([ct_2021_sum_all,ct_2022_sum_all])
df_all.min()
```

```python
pivot_all=pivot_all.fillna(0)
pivot_all
```

```python
pivot_all['monthly_avg']=pivot_all.mean(axis=1)
pivot_all.sort_values('monthly_avg',ascending=False)
```

```python
pivot_all.reset_index(inplace = True)
```

```python
ct_sel2=ct_sel[['BoroCT2020','NTAName','geometry']]
Data_2021_2022 = pivot_all.merge(ct_sel2, how="left",on='BoroCT2020')
Data_2021_2022
```

```python
newCols = []
for col in Data_2021_2022.columns:
    if type(col) == tuple:
        if col[1]:
            newCols.append(col[1])
        else:
            if 'BoroCT2020' == col[0]:
                newCols.append('BoroCT2020_1')
            else:
                newCols.append(col[0])
    elif type(col)==str:
        newCols.append(col)
    else:
        newCols.append(col)
        
Data_2021_2022.columns = newCols
```

```python
gdf_Data_2021_2022=gpd.GeoDataFrame(Data_2021_2022)
```
