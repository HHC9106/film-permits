```python
newTable = []

def getBoroCD(row):
    boroLookup = {
        'Manhattan': 100,
        'Brooklyn': 300,
        'Bronx': 200,
        'Queens': 400,
        'Staten Island': 500
    }
    
    boro = boroLookup[row['Borough']]
    
    if pd.isna(row['CommunityBoard(s)']):
        return;
        
    for cd in row['CommunityBoard(s)'].split(','):
        newTable.append({'BoroCD': int(cd) + boro, 'EventId': row['EventID']})
        
df.apply(getBoroCD, axis = 1)
```

```python
df_borocd_count = pd.DataFrame(newTable)['BoroCD'].value_counts()
df_borocd_count = df_borocd_count.reset_index()
df_borocd_count = df_borocd_count.rename(columns={'BoroCD':'counts','index':'BoroCD'})
df_borocd_count.head()
```

```python
df_borocd_count.to_csv(r'../data/permit_borocd_count.csv')
```
