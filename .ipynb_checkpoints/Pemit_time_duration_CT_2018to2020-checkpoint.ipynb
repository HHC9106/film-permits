```python
import warnings
warnings.filterwarnings("ignore")

import requests
import numpy as np
import pandas as pd
import geopandas as gpd # GeoPandas library for spatial analytics
from shapely.geometry import shape
from shapely.geometry import shape

from datetime import datetime, timedelta
```

```python
data = gpd.read_file("../output/film permit_city_from 2018.shp" )
```

```python
data.head(5)
```

```python
data_fil=data.drop(columns=['OBJECTID','Join_ID'])
data_fil=data_fil.drop_duplicates().reset_index()
```

```python
len(data_fil)/len(data)
```

```python
def range_to_list(a, b):
    return pd.date_range(a, b, freq = 'H')[:-1]
data_fil['StartEnd'] = [[i, j] for i, j in zip( data_fil['StartDateT'], data_fil['EndDateTim'])]
data_fil['TimeList'] = data_fil['StartEnd'].apply(lambda x: range_to_list(x[0], x[1]))
data_fil['hours'] = data_fil['TimeList'].apply(lambda x: list(x.hour))
data_fil['month'] = data_fil['TimeList'].apply(lambda x: list(set(x.month)))
data_fil['date'] = data_fil['TimeList'].apply(lambda x: list(set(x.date)))
data_fil['hour_sum'] = data_fil['hours'].apply(lambda x: len(x))  
data_fil['day_sum']=data_fil['hour_sum']/24
```

```python
data_fil
```

```python
data_fil['StartDateTime'] = data_fil['StartDateT'].apply(pd.to_datetime)
```

```python
data_2018 = data_fil[(data_fil['StartDateTime']>='2018-1-1') & (data_fil['StartDateTime']<'2019-1-1')]
```

```python
data_2018_buffer=data_2018
data_2018_buffer['geometry']= data_2018_buffer.buffer(3,cap_style=2)
```

```python
ct = gpd.read_file("../vector/2020 Census Tract_NYC" )
ct_sel = ct[['BoroCT2020','NTAName','GEOID','geometry']]
```

```python
ct_data_2018 = ct_sel.sjoin(data_2018_buffer, how="left")
ct_data_2018=ct_data_2018.dropna().reset_index()
```

```python
month1 = ct_data_2018.apply(lambda x: pd.Series(x['month']),axis=1).stack().reset_index(level=1, drop=True)
month1.name = 'month'
ct_data_2018 = ct_data_2018.drop('month', axis=1).join(month1)
ct_data_2018['month'] = ct_data_2018['month'].astype(str)
ct_data_2018.head(5)
```

```python
cd = gpd.read_file("../vector/Council Districts/nycc.shp")
cd33 = cd[cd['CounDist']==33]
```

```python
ct2018_cd33 = gpd.overlay(ct_data_2018, cd33, how='intersection')
ct2018_cd33.head(5)
```

```python
ct_2018_forsum=ct2018_cd33[['BoroCT2020','EventID','hour_sum','day_sum','month']]
ct_2018_forsum=ct_2018_forsum.drop_duplicates()
ct_2018_forsum['month']= ct_2018_forsum['month'].astype(float).astype(int)

ct_2018_forsum
```

```python
ct_2018_sum = ct_2018_forsum[['BoroCT2020','hour_sum','day_sum','month']].groupby(by=['BoroCT2020','month']).sum()
ct_2018_sum = ct_2018_sum.reset_index()
ct_2018_sum.sort_values('day_sum')
```

```python
ct_2018_sum[ct_2018_sum['month']=='nan']
```

```python
pivot_time_2018 = ct_2018_sum.pivot_table(index =['BoroCT2020'], values =['day_sum'],columns='month')
pivot_time_2018
```

```python
pivot_time_2018.to_csv('../output/Pivot_permits_cd33_2018_time.csv')
```
