# Per Month

```python
from datetime import date
```

```python
def range_to_list(a, b):
    return pd.date_range(a, b, freq = 'H')[:-1]
```

```python
data_sel_buffer['StartEnd'] = [[i, j] for i, j in zip( data_sel_buffer['StartDateT'], data_sel_buffer['EndDateTim'])]
data_sel_buffer['TimeList'] = data_sel_buffer['StartEnd'].apply(lambda x: range_to_list(x[0], x[1]))
data_sel_buffer['month'] = data_sel_buffer['TimeList'].apply(lambda x: list(set(x.month)))
```

```python
ct_sel2=ct_sel[['BoroCT2020','NTAName','geometry']]
```

```python
Data_2021_2022 = pivot_all.merge(ct_sel2, how="left",on='BoroCT2020')
Data_2021_2022
```

```python
newCols = []
for col in Data_2021_2022.columns:
    if type(col) == tuple:
        if col[1]:
            newCols.append(col[1])
        else:
            if 'BoroCT2020' == col[0]:
                newCols.append('BoroCT2020_1')
            else:
                newCols.append(col[0])
    elif type(col)==str:
        newCols.append(col)
    else:
        newCols.append(col)
        
Data_2021_2022.columns = newCols
```

```python
gdf_Data_2021_2022=gpd.GeoDataFrame(Data_2021_2022)
```
