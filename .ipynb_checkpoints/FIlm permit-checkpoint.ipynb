```python
cb_filters=['1','2','3','6']

def getFilter(row):
    if row['Borough'] != 'Brooklyn':
        return False
    
    cb_list =  row['CommunityBoard(s)'].split(',')
    for cb in cb_list:
        if cb.strip() in cb_filters:
            return True
    return False
  

df['Filter'] = df.apply(getFilter, axis = 1)
```

```python
import pickle
```

```python
# create the cache-like database to store the query.
#Use pickle to store the data into database that no longer need to rerun everytime

with open('oft_lookup.pkl', 'rb') as file:
    oft_lookup = pickle.load(file)

def getOFT(url):
    
    data = oft_lookup.get(url)
    if url not in oft_lookup:
        r = requests.get(url)
        data = r.json()
        oft_lookup[url] = data
        
    return data    
```

```python
segidlookup=[]

def getGeom(row):
    boro = row['Borough']
    
    segids = []
    for oft in row['OFT_Text']:
        oft_1 = oft.strip()
        try:
            on_street = oft_1.split('between')[0]
            from_street, to_street  = oft_1.split('between')[1].split('and')

            #print(on_street, from_street, to_street)

            url = f'https://www.locationservice.nyc/locationserviceapi//api/Block?OnStreet={on_street}&CrossStreetOne={from_street}&CrossStreetTwo={to_street}&Borough={boro}&BlockType=ExtendedStretch&ExtendedStretchType=Blockface&'
            data = getOFT(url)

            if 'BlockFaceList' in data:
                for i in data['BlockFaceList']:
                    if 'SegmentID' in i:
                        segids.append(i['SegmentID'])
                        segidlookup.append({
                            'EventID': row['EventID'],
                            'SegmentID': i['SegmentID']
                        })
        except:
            print(oft_1)
            continue
                    
    return segids
        #else:
           #print('\t',data['GeoSupportError'])
            

    #r = requests.get(f'https://www.locationservice.nyc/locationserviceapi//api/LionSegmentConflation?SegmentId={",".join(segids)}')
    
#     coords = []
#     all_segids = []
#     #print(r.json())
#     for i in r.json():
#         if 'GeomWebMercator' in i:
#             for arr in i['GeomWebMercator']['coordinates']:
#                 coords.append(arr)
#                 all_segids.append(i['NewSegmentId'])

#     newLineString = {'type': 'MultiLineString', 'coordinates': coords}
    
#     return pd.Series([shape(newLineString),all_segids])

      

df_film['segids'] = df_film.apply(getGeom, axis = 1)
```

```python
calendar = []
date_range = pd.date_range(start=datetime(2021,1,4), end=datetime(2022,3,8))
print(date_range)

for date in date_range:
    start_date = date
    end_date = date + timedelta(days=1)
    count = len(gdf_film_time[
        gdf_film_time['StartDateTime'].between (start_date, end_date) |
          gdf_film_time['EndDateTime'].between (start_date, end_date)])
    
    calendar.append({'date': date, 'count': count})
```

```python
permit_datecount = pd.DataFrame(calendar)
```

```python
permit_datecount
```

```python
Calendar_avg = []

for i in range(0,len(permit_datecount)):
    date = permit_datecount['date'][i]
    count = round(sum(permit_datecount['count'][i:i+6])/7,0)
    Calendar_avg.append({'date': date, 'count(avg)': count})
```

```python
permit_datecount_avg = pd.DataFrame(Calendar_avg)
```

```python
permit_count_all = permit_datecount.merge(permit_datecount_avg, how='left', on='date')
```

```python
permit_count_all
```

```python
permit_count_all.to_csv('../data/permit_in_date.csv',index=False)
```
