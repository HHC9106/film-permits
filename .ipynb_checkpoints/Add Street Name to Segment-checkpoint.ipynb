```python
import warnings
warnings.filterwarnings("ignore")

import requests
import numpy as np
import pandas as pd
import geopandas as gpd # GeoPandas library for spatial analytics
from shapely.geometry import shape
from shapely.geometry import shape
```

```python
lion = gpd.read_file("../vector/lion_line" )
```

```python
#unlimit columuns display
pd.set_option('display.max_columns', None)
```

```python
lion_name = lion[['Street','SegmentID']]
lion_name
```

```python
df_geo=gpd.read_file('../vector/Segment count_GeoBoundary/Segment count_GeoBoundary.shp')  
df_geo = df_geo[['SegmentID','counts','CounDist','boro_cd','geometry']]
df_geo.set_index('SegmentID')
```

```python
df_seg_name = df_geo.merge(lion_name,how='left',on='SegmentID')
df_seg_name.rename(columns={"counts": "Counts","counts": "Counts"})
df_seg_name.head(2)
```

```python
df_seg_name = df_seg_name.rename(columns={"counts":"Counts","CounDist":"Council District","boro_cd":"Community District"})
df_seg_name = df_seg_name.set_index('SegmentID')
df_seg_name
```

```python
df_seg_name['Community District'] = round(df_seg_name['Community District']).astype(int)
df_seg_name['Counts'] = df_seg_name['Counts'].astype(int)
df_seg_name
```

```python
df_seg_name.to_file('../output/permit_city_segment_count.shp')  
```
